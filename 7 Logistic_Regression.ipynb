```python
from sklearn.impute import SimpleImputer
```

```python
simple_imputer=SimpleImputer(strategy='mean')
```

```python
df['Age']=simple_imputer.fit_transform(df[['Age']])
```

```python
simple_imputer1=SimpleImputer(strategy='most_frequent')
```

```python
#df['Embarked']=simple_imputer1.fit_transform(df[['Embarked']])
df['Embarked'] = simple_imputer1.fit_transform(df['Embarked'].values.reshape(-1, 1))[:, 0]
```

```python
df=df.drop('Cabin',axis=1)
```

```python
df.head()
```

```python
df['Embarked'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
df=pd.get_dummies(data=df,columns=['Sex','Embarked'],dtype='int')
```

```python
df.head()
```

```python
target=df['Survived']
```

```python
features=df.drop(['Survived','Name','Ticket','PassengerId'],axis=1)
features
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.80,random_state=100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,f1_score,classification_report,roc_auc_score,roc_curve,confusion_matrix
logistic=LogisticRegression()
```

```python
logistic.fit(x_train,y_train)
```

```python
import joblib
joblib.dump(logistic,'model.pkl')
model = joblib.load('model.pkl')
```

```python
y_pred=logistic.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
pred_prob=logistic.predict_proba(x_test)[:,1]
pred_prob
```

```python
target.value_counts()
```

```python
AUC=roc_auc_score(y_test,pred_prob)
AUC
```

```python
fpr,tpr,thr=roc_curve(y_test,pred_prob)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label='Logistic Regression (area = {:.2f})'.format(AUC))
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='best')
plt.grid(True)
```

## 1. What is the difference between precision and recall?
   Precision is the percentage of correct positive predictions out of all predicted positives.
   Recall is the percentage of actual positives correctly identified.
   Precision focuses on avoiding false positives, while recall focuses on finding all true positives.
   We use precision when accuracy matters and recall when catching all positives is critical.

## 2. What is cross-validation, and why is it important in binary classification?
   Cross-validation splits data into parts to train and test the model multiple times.
   It helps prevent overfitting and gives a more accurate performance estimate.
   By testing on different subsets, the model learns to generalize better.
