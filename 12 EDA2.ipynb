## Applying Scaling Methods

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
std_sca=StandardScaler()
standard_scaled_ncolumns=df.copy()
```

```python
standard_scaled_ncolumns[['age','fnlwgt','education_num','capital_gain','capital_loss','hours_per_week']]=std_sca.fit_transform(df[['age','fnlwgt','education_num','capital_gain','capital_loss','hours_per_week']])
```

```python
standard_scaled_ncolumns.head()
```

```python
min_max_sca=MinMaxScaler()
min_max_sca_ncolumns=df.copy()
```

```python
min_max_sca_ncolumns[['age','fnlwgt','education_num','capital_gain','capital_loss','hours_per_week']]=min_max_sca.fit_transform(df[['age','fnlwgt','education_num','capital_gain','capital_loss','hours_per_week']])
```

```python
min_max_sca_ncolumns
```

## Encoding Techniques

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df_one_hot=pd.get_dummies(df,columns=['sex','workclass'],dtype='int')
df_one_hot.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
```

```python
for column in ['workclass','education', 'marital_status', 'occupation', 'relationship', 'race', 'native_country']:
    df[column] = label_encoder.fit_transform(df[column])
```

```python
df.head()
```

## Feature Engineering

```python
df['work_per_year'] = df['hours_per_week'] * 52
df['work_per_year']
```

```python
df['age_group']=pd.cut(df['age'],bins=(0,18,35,55,100),labels=('0-18','18-35','35-55','55-100'))
```
