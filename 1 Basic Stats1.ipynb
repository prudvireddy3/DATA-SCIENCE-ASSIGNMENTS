### ‚óè	Boxplots:
	Create boxplots for numerical variables to identify outliers and the interquartile range.
	Discuss any findings, such as extreme values or unusual distributions.


```python
fig = plt.figure(figsize =(10, 7))
for i in num_col:
    df.boxplot(column=num_col); 
plt.show()
```

```python
Q1=df[num_col].quantile(0.25)
Q3=df[num_col].quantile(0.75)
IQR=Q3-Q1
print(Q1,Q3,IQR)
```

```python
lower_qua=Q1-(1.5*IQR)
upper_qua=Q3+(1.5*IQR)
print(lower_qua,upper_qua)
```

```python
df_outliers=df[(df[num_col]<lower_qua) | (df[num_col]>upper_qua)]
df_outliers
```

```python
df_no_outliers=df[(df[num_col]>lower_qua) & (df[num_col]<upper_qua)]
df_no_outliers
```

```python
figure=plt.figure(figsize=(12,6))
df_no_outliers.boxplot()
plt.show()
```

### Standardization of Numerical Variables

```python
num_col=[i for i in df.columns if df[i].dtype!='O']
```

```python
num_col
```

```python
fig, axs = plt.subplots(nrows=len(num_col), ncols=2, figsize=(15, 5 * len(num_col)))
fig.suptitle('Before and After Standardization', fontsize=20)

for i, column in enumerate(num_col):
    sns.histplot(df[column], ax=axs[i, 0], kde=True)
    axs[i, 0].set_title(f'{column} - Before Standardization')
    
for column in num_col:
    mean = df[column].mean()
    std = df[column].std()
    standardised_data = (df[column] - mean) / std
    
for i, column in enumerate(num_col):
    sns.histplot(standardised_data, ax=axs[i, 1], kde=True)
    axs[i, 1].set_title(f'{column} - After Standardization')
    
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### Conversion of Categorical Data into Dummy Variables

```python
from sklearn.preprocessing import OneHotEncoder
one_hot=OneHotEncoder(sparse=False,drop='first')
one=(one_hot.fit_transform(df[cat_col]))
df1=pd.DataFrame(one,columns=one_hot.get_feature_names_out(cat_col))
df2=pd.concat([df.drop(columns=cat_col),df1],axis=1)
print(df2.head())
```
