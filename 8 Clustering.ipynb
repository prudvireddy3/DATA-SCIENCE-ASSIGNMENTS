### K-Means
## Elbow_method

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.metrics import silhouette_score
```

```python
inertia=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_new)
    inertia.append(kmeans.inertia_)
print(inertia)    
```

```python
plt.plot(range(1,11),inertia)
plt.xlabel('No_Of_Clusters')
plt.ylabel('inertia')
plt.title('Elbow_method',size=20)
```

```python
output=[]
for i in range(2,11):
    labels=KMeans(n_clusters=i).fit(df_new).labels_
    score=silhouette_score(df,labels)
    output.append([i,score])
print(output)    
    
```

```python
cluster=KMeans(n_clusters=2)
labels=cluster.fit(df_new).labels_
```

### Hiererchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_new,method='ward'))
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_new,method='average'))
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_new,method='single'))
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_new,method='complete'))
```

```python
output=[]
for i in range(2,11):
    labels=AgglomerativeClustering(n_clusters=i,linkage='complete').fit(df_new).labels_
    score=silhouette_score(df_new,labels)
    output.append([i,score])
print(output)    
```

```python
agg=AgglomerativeClustering(n_clusters=2,linkage='complete',affinity = 'euclidean')
```

```python
cluster=agg.fit(df_new)
```

```python
hierarchy_labels=cluster.labels_
```

```python
silhouette_score(df_new,cluster.labels_)
```

```python
df['hierar_labels']=hierarchy_labels
```

### DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
output=[]
eps=[0.1,0.2,0.5,0.8,1.2,1.6]
min_samples=range(1,11)
for i in eps:
    for j in min_samples:
        labels=DBSCAN(eps=i,min_samples=j).fit(df_new).labels_
        score=silhouette_score(df_new,labels)
        output.append([i,j,score])
print(output)        
        
```
